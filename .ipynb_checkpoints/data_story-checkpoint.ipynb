Having cleaned the data, the next step is to visualize the data to investigate any trends and find some insights about the data. I will be focusing on the combined dataset created from the Cleveland, Hungary, Switzerland, and Long Beach heart disease data. I am primarily interested in finding any relationships between the columns of the data, particularly any correlations between variables and whether the patient has heart disease. The columns of the combined dataset are:

+ age
+ sex (0 = female, 1 = male)
+ cp (chest pain type; 1 = typical angina, 2 = atypical angina, 3 = non-anginal pain, 4 = asymptomatic)
+ trestbps (resting blood pressure in mm Hg)
+ chol (serum cholesterol in mg/dl)
+ restecg (resting electrocardiographic results; 0 = normal, 1 = ST-T wave abnormality, 2 = probable/definite left ventricular hypertrophy)
+ thalach (maximum heart rate)
+ exang (exercise induced angina; 0 = false, 1 = true)
+ oldpeak (ST depression induced by exercise relative to rest)
+ num (heart disease diagnosis; 0 = absence, 1-3 = presence)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('combined_data.csv', index_col=0)

df.head()
```

I first construct a pair plot using seaborn to investigate any possible trends or correlation between columns of the data. I specify the hue as the heart disease predictor since that is what I am primarily interested in comparing.

```python
%matplotlib inline
sns.pairplot(df, hue='num')
```

There are a number of interesting correlations in these plots:

+ Men are more at risk of heart disease than women
+ Patients with asymptomatic chest pain (cp = 4) are more likely to have heart disease than patients with other types of chest pain
+ There seems to be a negative correlation between age and maximum heart rate (thalach)
+ Interestingly, patients with higher maximum heart rates seem to be less at risk of heart disease
+ Patients with exercise-induced angina (exang = 1) are much more likely to have heart disease
+ Older patients seem more at risk of heart disease than younger ones

First, I will investigate the difference in heart disease rates between men and women.

```python
df['sex'] = df['sex'].map({0:'female', 1:'male'})
```

```python
g = sns.countplot(y='sex', hue='num', orient='h', data=df)
plt.legend(title='heart disease', labels=['no', 'yes'])
plt.show(g)
```

```python
table = df.pivot_table(values='age', index='sex', columns='num', aggfunc='count')
table
```

From the graph and table, it is clear that men in this dataset are much more likely to have heart disease than women. In fact, men are nearly 2.5 times as likely to have heart disease than women. This agrees with medical studies that have found that men are more at risk of heart disease than women.

Next, I will visualize the relation between type of chest pain and heart disease.

```python
g = sns.countplot(y='cp', hue='num', orient='h', data=df.sort_values(by='cp'))
g.set(yticklabels=['typical', 'atypical', 'non-anginal pain', 'asymptomatic'])
plt.ylabel('angina type')
plt.legend(title='heart disease', labels=['no', 'yes'])
plt.show(g)
```

```python
table = df.pivot_table(values='age', index='cp', columns='num', aggfunc='count')
table
```

Interestingly, the data suggests that patients without any type of pain (asymptomatic) are much more likely to have heart disease than patients with chest pain or other types of pain. This seems to contradict the fact that angina is a common symptom of cardiovascular problems. One possible hypothesis to explain this phenomenon is that perhaps many patients admitted to the hospital for heart disease are in critical condition and unable to answer questions about chest pain, leading to medical professionals to simply document them as being 'asypmtomatic'. Verifying this would require contacting the people who collected this data, which could be difficult considering how old this study is.

Next up is investigating the correlation between exercise-induced angina and heart disease.

```python
g = sns.countplot(y='exang', hue='num', orient='h', data=df.sort_values(by='exang'))
g.set(yticklabels=['no info', 'false', 'true'])
plt.ylabel('exercise-induced angina')
plt.legend(title='heart disease', labels=['no', 'yes'])
plt.show(g)
```

```python
table = df.pivot_table(values='age', index='exang', columns='num', aggfunc='count')
table
```

From the data, 36% of patients without exercise-induced angina had heart disease, while 84% of patients with exercise-induced angina had heart disease. This is a huge difference, and the presence of exercise-induced angina in patients could be one of the biggest predictors of heart disease.

Next I will look at the relation between age and maximum heart rate in patients.

```python
g = sns.jointplot(x='age', y='thalach', data=df, kind='reg')
plt.ylabel('maximum heart rate')
plt.show(g)
```

```python
new_df = df
new_df['num'] = new_df['num'].map({0:'no', 1:'yes'})
new_df = new_df.rename(columns={'num':'heart disease'})
```

```python
g = sns.residplot(x='age', y='thalach', data=new_df)
plt.ylabel('maximum heart rate')
plt.show(g)
```

```python
g = sns.lmplot(x='age', y='thalach', data=new_df, hue='heart disease')
plt.ylabel('maximum heart rate')
plt.show(g)
```

From these plots, there is a clear negative correlation between age and maximum heart rate in patients, which is consistent with medical studies. The color-coded scatterplot also seems to suggest that older patients with lower heart rates are more likely to have heart disease, as there are many more green points there than blue points.

Next I will look at the heart rate distribution of patients with and without heart disease.

```python
dist = [new_df[new_df['heart disease'] == i]['thalach'] for i in ('no', 'yes')]
```

```python
fig, ax = plt.subplots()
sns.distplot(dist[0], ax=ax, kde=False, norm_hist=True, label='no heart disease')
sns.distplot(dist[1], ax=ax, kde=False, norm_hist=True, label='heart disease')
plt.xlabel('maximum heart rate')
plt.legend()
plt.show()
```

Something very interesting here is that patients with higer maximum heart rates are less likely to have heart disease than patients with lower heart rates, which seems to contradict the fact that high heart rates can be a symptom of heart disease. However, this is likely explained by the fact that patients with higher heart rates are typically younger (as explored in the age vs. heart rate plots), and older patients are more likely to develop heart disease than younger patients. I examine this further by plotting the age distribution of patients with and without heart disease:

```python
dist = [new_df[new_df['heart disease'] == i]['age'] for i in ('no', 'yes')]
```

```python
fig, ax = plt.subplots()
sns.distplot(dist[0], ax=ax, kde=False, norm_hist=True, label='no heart disease')
sns.distplot(dist[1], ax=ax, kde=False, norm_hist=True, label='heart disease')
plt.legend()
plt.show()
```

This distribution does indeed suggest that older patients are at higher risk of heart disease than younger ones, which can explain the interesting correlation between heart rate and heart disease.

From the data exploration here, it becomes clear how some of these variables are correlated to the presence of heart disease. Building a model using all of these features combined will likely be effective in predicting the presence of heart disease in patients.
